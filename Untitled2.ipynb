```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5]
y=[3,7,8,5,10]
plt.plot(x,y,linestyle='--',color='r')
#plt.show()
```

```python
x=['JAN','FEB','MAR','APR','MAY','JUN']
y=[2000,3000,1000,4000,2000,6000]
plt.plot(x,y,linestyle='--',marker='o')
plt.title("SAVING OF 6 MONTHS")
plt.xlabel("months")
plt.ylabel("amount")
plt.show()
```

```python
x=['JAN','FEB','MAR','APR','MAY','JUN']
y=[2000,3000,1000,4000,2000,6000]
plt.bar(x,y,color=['pink','r'],edgecolor='black')
plt.title("BAR PLOT")
plt.xlabel("months")
plt.ylabel("amount")
plt.show()
```

```python
x=[10,20,30,10,20,30]
#y=[2000,3000,1000,4000,2000,6000]
plt.hist(x,bins=6,color=['pink'],edgecolor='black')
plt.title("hist PLOT")
plt.xlabel("months")
plt.ylabel("amount")
plt.show()
```

```python
x=['JAN','FEB','MAR','APR','MAY','JUN']
y=[2000,3000,1000,4000,2000,6000]
plt.pie(y,labels=x,colors=('pink','r','lightpink','lightblue','black','orange'),)
plt.title("pie PLOT")
plt.xlabel("months")
plt.ylabel("amount")
plt.legend()
plt.show()
```
