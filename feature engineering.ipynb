```python
#Extract Date & Time Features

import pandas as pd

#Sample dataset

df=pd.DataFrame({'TransactionDate': pd.to_datetime(['2025-02-05 14:30:00', '2025-02-06 16:45:00'])})

#Extract date-related features

df ['DayOfWeek'] = df['TransactionDate'].dt.dayofweek #Monday=0, Sunday=6
df ['Hour']=df['TransactionDate'].dt.hour # Extract hour
df['Isleekend'] = df['DayOfWeek'].apply(lambda x: 1 if x>= 5 else 0)

# Weekend flag

print(df)

#vwhy? Helps capture behavioral trends (e.g., shopping habits on weekends vs. weekdays).
```

```python
#Aggregated Features
#rind average transaction amount per user:

df_transactions=pd.DataFrame({

'UserID': [101, 102, 101, 103, 102],
'TransactionAmount': [500, 300, 700, 1000, 400]
})

df_user_avg=df_transactions.groupby('UserID')['TransactionAmount'].mean().reset_index()
df_user_avg.rename(columns={'TransationAmount': 'AvgTransactionAmount'}, inplace=True)

print(df_user_avg)

#Identifies high-value customers and spending patterns.
```
