```python
import pandas as pd
gt=pd.read_csv(r"C:\Users\CVR\Downloads\GlobalLandTemperaturesByCity.csv.zip")
gt
```

```python
gt.isna().sum()
```

```python
gt.dtypes
```

```python
mean=gt['AverageTemperature'].mean()
#mode_value=mode.iloc[0]
mean
```

```python
import numpy as np
gt['AverageTemperature']=gt['AverageTemperature'].replace(np.nan,mean)
gt
```

```python
mean=gt['AverageTemperatureUncertainty'].mean()
#mode_value=mode.iloc[0]
mean
```

```python
import numpy as np
gt['AverageTemperatureUncertainty']=gt['AverageTemperatureUncertainty'].replace(np.nan,mean)
gt
```

```python
gt['dt']=pd.to_datetime(gt['dt'])
gt.dtypes
```

```python
gt['year']=gt['dt'].dt.year
gt
```

```python
annual_avg_temp=gt.groupby('year')['AverageTemperature'].mean().reset_index()
annual_avg_temp
```

```python
import matplotlib.pyplot as py  
```

```python
py.plot(annual_avg_temp['year'],annual_avg_temp['AverageTemperature'])
py.title("AVERAGE TEMPERATURE ACROSS YEARS")
py.show
```

```python
avg_city_temp=gt.groupby('Country')['AverageTemperature'].mean().reset_index()
avg_city_temp
```

```python
avg_city_temp = avg_city_temp.sort_values(by='AverageTemperature', ascending=False)
avg_city_temp
```
