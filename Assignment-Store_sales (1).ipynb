# Assignment on Store_sales dataset

**Load the Dataset**

```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\Lenovo\Downloads\store_sales.csv")
df
```

**Calculate the total sales for each store across all months.**

```python
#totsales=df.groupby('city')[[['Jan']['Feb']['Mar']['Apr']['May']['Jun']['July']['Aug']['Sep']['Oct']['Nov']['Dec']]].sum()
#otsales

totsales = df.groupby('store_id')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                               'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum().sum(axis=1)
print(totsales)
#sum(axis=1) then we get summed up total 
```

**Find the average sales for each month across all stores**

```python
#avgsales = df.[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                              # 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean(axis=1)
#print(avgsales)

# Calculate the average sales per store across all months
df['Average Sales'] = df[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean(axis=1)

# Display the store ID and their corresponding average sales
print(df[['store_id', 'Average Sales']])
```

**Identify the store with the highest total sales.**

```python
strid = df.groupby('storeid')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                               'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum()
strid
```

**Calculate the total sales for each city.**

```python
tot=df.groupby('city')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                               'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum().sum(axis=1)
tot
```

**List stores with total sales greater than 200**

```python
totgt200 = df[df[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum(axis=1)>200]
totgt200
```
